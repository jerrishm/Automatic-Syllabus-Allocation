```python
import pandas
from pandas import read_csv
import numpy
from keras.models import Sequential
from keras.layers import Dense
```

```python
dataframe = pandas.read_csv("Desktop/Datasets/test1.csv")
dataset = dataframe.values
```

```python
x = dataset[:,1:21]
```

```python
y = dataset[:,21]
```

```python
y
```

```python
model = Sequential()
model.add(Dense(20, input_dim=20, activation='relu'))
model.add(Dense(1))
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(dataset[:,1:21], dataset[:,21], test_size=0.25, random_state=87)
```

```python
model.fit(x_train,y_train,nb_epoch =100, batch_size =16,verbose =2)
```

```python
prediction = model.predict(x).astype('int')
```
